#### Questions 3: Types of Data

- Briefly explain the concepts of structured, semi-structured and unstructured data.
- Give an example of each type of data.

_Answer:_ 

**Structured data** is the data type that is the easiest to search and organize because it is contained in rows and columns and its elements can be mapped into fixed pre-defined fields. An example of structured data is financial data including accounting transactions, address details, demographic information etc.

**Semi-structured data** has some defining or consistent characteristics but doesn’t conform to a structure as rigid as is expected with a relational database; A good example of semi-structured data is email messages.

**Unstructured data** is data that cannot be contained in a row-column database and doesn’t have an associated data model; An example of unstructured is text.

#### Questions 6: Declarative versus Procedural Languages

- Briefly explain the difference between a declarative data manipulation language and a procedural data manipulation language.
- Briefly explain why relational algebra is procedural and Python is declarative.

_Answer:_ 

A procedural data manipulation language is a set of commands that are written in order. Programming languages are considered good examples of it, like C, java, etc. They consist of a set of operations that take several relations as an input and produce a new relation as an output.

A declarative data manipulation language is a language that is written using only the promised definition; It focuses only on the logic of the program without describing how it actually flows.

Relational Algebra refers to a procedural language that describes the desired information and how to derive that information in a relational DB. Therefore, relational algebra, as we described, is a procedural language because it describes how to derive the desired information using operators and computational rules.

On the other hand, Python is not just one of them, but a programming language that properly mixes the two; Python can also implement functions as a procedural language that is executed in order with several operations and variables. Python also allows us to define and translate the data we want through various libraries effectively on the database schema as a declarative language 

#### Questions 7: Application Architectures

- Briefly explain the concepts of two-tier and three-tier application architectures.
- Are Jupyter notebooks two-tier or three-tier applications?

_Answer:_

A two-tier architecture consists of two layers: the client layer and the database layer.

It refers to a form in which a web server is connected to a DB. The 2-tier structure is a commonly used structure when there is not much traffic. A 2-tier-based client/server environment is the most common method of application development, in which one server process is created on one client.

The three-tier architecture is the most popular implementation of a multi-tier architecture and consists of a single presentation tier(user interface), logic tier, and data tier; This architecture is used to run client-server applications such as front-end, back-end, and independently managed databases with specific tasks at each layer.

Jupyter notebooks are three-tier applications. One of the differences between a two-tier structure and a three-tier structure is security. Because a three-tier structure does not allow the presentation layer and the data layer to communicate directly, a well-designed application layer can work as a type of internal firewall to prevent SQL injection and other malicious behavior.

#### Questions 9: Relational Algebra

- Briefly explain the concept of the relational algebra being _closed_ under the operations/operators.
- Give an example.

_Answer:_

The relational algebra is a theoretical procedural query language which takes an instance of relations and does operations that work on one or more relations to describe another relation without altering the original relation(s). Thus, both the operands and the outputs are relations. So the output from one operation can turn into the input to another operation, which allows expressions to be nested in the relational algebra, just as nest arithmetic operations.

Example:

π student.ID, age(σ name  = 'KeunWoo'(CUstudent))

Starting with all rows from CUstudent whose attribute name is "KeunWoo." Since the relational algebra being closed under the operations/operators, returning value should be another relation. Then we utilize the new relation with other standards, such as student ID, age (among whose name is "KeunWoo") to get the final result.
